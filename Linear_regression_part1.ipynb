Step1 Gathering Dataset

```python
# Loading and reading the dataset
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance (2).csv")
```

```python
data.head()
```

```python
data.info()
```

Step2- Data cleaning/Data preprocessing /Data Wrangling

```python
#one hot encoding will be done to convert categorical variable to numerical variables.
data['sex']=data['sex'].apply({'male':0,'female':1}.get)

```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'no':0,'yes':1}.get)

data['region']=data['region'].apply({'southeast':0,'southwest':1,'northeast':2,'northwest':3}.get)
```

```python
data.head()
```

Step3-Divide the data into dependent(y) and independent(x) variables

```python
x=data[['age','sex','bmi','children','smoker','region']] # thease are the independent variables
y=data['charges'] #this is the dependent variable
```

```python
x.head()
```

```python
y.head()
```

Step4- Splitting the data into training and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) #test_size=0.2 means 20% of the data will be used for testing and remaining 80% will be used for data for training the machine learning model
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

Creating a machine learning model by Using linear regression algorithm

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression() #Linear regression algorith gets activated
```

```python
regression.fit(x_train,y_train) #fit() is used to train the machine learning model by providing x_train and y_test as input variables
```

```python
predictions=regression.predict(x_test) # predict is used to predict the output from machine learing model which then gets compared y_test
```

```python
predictions #chargest predicted by machine learing model (predicted values)
```

```python
regression.score(x,y)
```
