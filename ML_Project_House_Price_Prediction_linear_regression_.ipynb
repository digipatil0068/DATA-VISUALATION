Gathering Data

```python
import numpy as np
import pandas as pd
```

```python
data1=pd.read_csv("/content/Housing (1).csv")
```

```python
data1.head()
```

```python
data1.info()
```

data cleaning

```python
data1['mainroad']=data1['mainroad'].apply({'no':0,'yes':1}.get)
```

```python
data1.head()
```

```python
data1['guestroom']=data1['guestroom'].apply({'no':0,'yes':1}.get)
data1['basement']=data1['basement'].apply({'no':0,'yes':1}.get)
data1['hotwaterheating']=data1['hotwaterheating'].apply({'no':0,'yes':1}.get)
data1['airconditioning']=data1['airconditioning'].apply({'no':0,'yes':1}.get)
data1['prefarea']=data1['prefarea'].apply({'no':0,'yes':1}.get)
data1['furnishingstatus']=data1['furnishingstatus'].apply({'furnished':1,'unfurnished':0,'semi-furnished':2}.get)
```

```python
data1.head()
```

divide the data into x variable independent and y varibale dependent variable

```python
x=data1[['airconditioning','area','bathrooms','stories','mainroad','guestroom','hotwaterheating','bedrooms','parking','prefarea','furnishingstatus']]
y=data1['price']
```

```python
x.head()
```

```python
y.head()
```

spliting the data into training and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

creating ml model by using linear regression algorithm

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression() #model activated
```

```python
regression.fit(x_train,y_train)
```

```python
pridictions=regression.predict(x_test)
```

```python
pridictions
```

```python
regression.score(x,y)
```
